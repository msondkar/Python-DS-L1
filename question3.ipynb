```python
import pandas as pd
from pandas import datetime
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline 

import warnings
warnings.simplefilter(action='ignore')
```

```python
# Load daily/weekly Retail prices of Onion

def parser(x):
    return datetime.strptime(x, '%d-%m-%y')

data = pd.read_csv('data/daily_retail_price_Onion-upto_apr_2015.csv', index_col=0, 
                   parse_dates=[0], date_parser=parser)
```

```python
print("Total number of Market Centres :", data['Centre_Name'].nunique())
```

```python
# name of market centres
data['Centre_Name'].unique()
```

```python
def round_mean(x):
    return round(np.mean(x), 2)

# find out minimum, maximum and average onion price by market center.
centre_data = data.groupby(['Centre_Name'])['Price'].agg({'Minimum_Price' : 'min', 'Maximum Price' : 'max', 
                                            'Average Price' : round_mean})
```

```python
centre_data
```

```python
# 10 market centers with maximum onion prices
centre_data['Maximum Price'].sort_values(ascending=False).head(10)
```

```python
# find out onion price trends for metro cities
metro_cities = ['MUMBAI', 'DELHI', 'KOLKATA', 'BENGALURU', 'CHENNAI', 'HYDERABAD']

fig, ax = plt.subplots(1, 1, figsize=(18,6))

for centre in metro_cities:
    centre_df = data[data.Centre_Name == centre]['Price']
    ax.plot(centre_df.index, centre_df.values, label=centre)

ax.legend(loc='best')
ax.grid(True, ls='--')
```

Onion prices tend to follow similar trends across metro cities.
